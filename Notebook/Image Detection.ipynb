**Cell 4 (Consolidate & Fix Labels)**

```python
# Class mapping
CLASS_MAPPING = {
    'mosquito': 0,
    'termite': 1,
    'cockroach': 2,
    'rodent': 3
}

# Create consolidated structure
CONSOLIDATED_DIR = 'consolidated_dataset'
os.makedirs(f"{CONSOLIDATED_DIR}/images/train", exist_ok=True)
os.makedirs(f"{CONSOLIDATED_DIR}/images/val", exist_ok=True)
os.makedirs(f"{CONSOLIDATED_DIR}/labels/train", exist_ok=True)
os.makedirs(f"{CONSOLIDATED_DIR}/labels/val", exist_ok=True)

def remap_and_fix_labels(label_file, new_class_id):
    """Remap class IDs and fix formatting"""
    with open(label_file, 'r') as f:
        lines = f.readlines()

    fixed_lines = []
    for line in lines:
        line = line.strip()
        if line:
            parts = line.split()
            if len(parts) >= 5:
                try:
                    # Validate and fix
                    x, y, w, h = map(float, parts[1:5])
                    fixed_lines.append(f"{new_class_id} {x} {y} {w} {h}\n")
                except ValueError:
                    continue  # Skip invalid lines

    return fixed_lines

# Process datasets
total_train = 0
total_val = 0

for ds in downloaded_datasets:
    pest_class = ds['class']
    new_class_id = CLASS_MAPPING[pest_class]
    dataset_path = ds['path']

    print(f"\nğŸ”„ Processing {ds['name']} â†’ Class {new_class_id} ({pest_class})")

    # Process train
    train_img_src = f"{dataset_path}/train/images"
    train_lbl_src = f"{dataset_path}/train/labels"

    if os.path.exists(train_img_src):
        for img_file in os.listdir(train_img_src):
            img_name = f"{ds['name']}_{img_file}"
            shutil.copy(f"{train_img_src}/{img_file}", f"{CONSOLIDATED_DIR}/images/train/{img_name}")

            label_file = img_file.replace('.jpg', '.txt').replace('.png', '.txt')
            if os.path.exists(f"{train_lbl_src}/{label_file}"):
                fixed_labels = remap_and_fix_labels(f"{train_lbl_src}/{label_file}", new_class_id)
                if fixed_labels:
                    with open(f"{CONSOLIDATED_DIR}/labels/train/{img_name.replace('.jpg', '.txt').replace('.png', '.txt')}", 'w') as f:
                        f.writelines(fixed_labels)
                    total_train += 1

    # Process val
    val_img_src = f"{dataset_path}/valid/images"
    val_lbl_src = f"{dataset_path}/valid/labels"

    if os.path.exists(val_img_src):
        for img_file in os.listdir(val_img_src):
            img_name = f"{ds['name']}_{img_file}"
            shutil.copy(f"{val_img_src}/{img_file}", f"{CONSOLIDATED_DIR}/images/val/{img_name}")

            label_file = img_file.replace('.jpg', '.txt').replace('.png', '.txt')
            if os.path.exists(f"{val_lbl_src}/{label_file}"):
                fixed_labels = remap_and_fix_labels(f"{val_lbl_src}/{label_file}", new_class_id)
                if fixed_labels:
                    with open(f"{CONSOLIDATED_DIR}/labels/val/{img_name.replace('.jpg', '.txt').replace('.png', '.txt')}", 'w') as f:
                        f.writelines(fixed_labels)
                    total_val += 1

print(f"\nâœ… Dataset Consolidation Complete!")
print(f"   ğŸ“Š Training images: {total_train}")
print(f"   ğŸ“Š Validation images: {total_val}")
print(f"   ğŸ“Š Total: {total_train + total_val}")
```

**ğŸš€ Cell 7: Train YOLOv8 (WITH AUTO-SAVE & RESUME)**

```python
from ultralytics import YOLO
import os
# Check if we have a checkpoint to resume from
checkpoint_path = f'{PROJECT_DIR}/eastern_pest_detection/weights/last.pt'
if os.path.exists(checkpoint_path):
    print(f"âœ… Found checkpoint! Resuming training from: {checkpoint_path}")
    model = YOLO(checkpoint_path)
    resume = True
else:
    print("ğŸ†• Starting fresh training with YOLOv8s")
    model = YOLO('yolov8s.pt')
    resume = False
# Train with auto-save every 5 epochs
results = model.train(
    data='data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='eastern_pest_detection',
    project=PROJECT_DIR,
    patience=10,
    save=True,
    save_period=5,  # Save checkpoint every 5 epochs
    device=0,
    workers=2,
    verbose=True,
    resume=resume  # Resume if checkpoint exists
)
print("\nâœ… Training Complete!")
print(f"   ğŸ“ Model saved to: {PROJECT_DIR}/eastern_pest_detection/weights/best.pt")
```

**ğŸ“Š Cell 7: Validate Model**

```python
# Load best model
best_model = YOLO(f'{PROJECT_DIR}/eastern_pest_detection/weights/best.pt')
# Validate
metrics = best_model.val()
print("\nğŸ“Š Final Model Performance:")
print(f"   mAP50: {metrics.box.map50:.3f}")
print(f"   mAP50-95: {metrics.box.map:.3f}")
print(f"   Precision: {metrics.box.mp:.3f}")
print(f"   Recall: {metrics.box.mr:.3f}")
if metrics.box.map50 > 0.85:
    print("\nğŸ‰ Excellent accuracy! Model is ready for deployment!")
elif metrics.box.map50 > 0.75:
    print("\nâœ… Good accuracy! Model is usable.")
else:
    print("\nâš ï¸ Low accuracy. Consider training longer or checking data quality.")
```
